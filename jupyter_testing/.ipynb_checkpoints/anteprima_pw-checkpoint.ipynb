```python
# Unisci i dati tra le tabelle products e categories
merged_data = products.merge(categories, how="left", left_on="fk_category", right_on="pk_category")

# Calcola la quantità venduta per ciascuna categoria
product_category_sales = merged_data["name"].value_counts()  # Usa il nome della categoria

# Crea il grafico a barre
plt.figure(figsize=(20,10))
sns.barplot(x= product_category_sales.index, y=product_category_sales.values)
plt.title("Best Selling Product Categories")
plt.xlabel("Product Categories")
plt.ylabel("Quantity Sold")
plt.xticks(rotation=45)  # Ruota le etichette sull'asse x se necessario
plt.show()
```
